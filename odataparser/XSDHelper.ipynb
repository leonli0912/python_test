```python
import xml.etree.ElementTree as ET

class XSDHelper(object):
    
    def __init__(self,path):
        print('init...')
        self._tree = ET.parse(path)
        self._root = self._tree.getroot() 
        print(self._root)
    def print_child(self,father):
        print(father.tag)
        for child in father:
            if child.tag == '{http://schemas.microsoft.com/ado/2008/09/edm}EntityType':
                #EntityType => complexType
                print('handle entity')
            print(child.tag)
            
            self.print_child(child)
            
if __name__ == '__main__':
    PATH = 'metadata.XML'
    xsdHelper = XSDHelper(PATH)
    xsdHelper.print_child(xsdHelper._root)
else:
    print('not in main program')
```
